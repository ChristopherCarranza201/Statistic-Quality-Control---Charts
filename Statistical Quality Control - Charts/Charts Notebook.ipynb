```python
import pandas as pd
import matplotlib.pyplot as plt

dfs = pd.read_excel("grafico s.xlsx")
dfx = pd.read_excel("grafico x.xlsx")
dfp = pd.read_excel("grafico p.xlsx")
dfr = pd.read_excel("grafico r.xlsx")
```

```python
x = dfs["Subgrupo"]
y1 = dfs["S"]
y2 = dfs["LSC"]
y3 = dfs["LC "]
y4 = dfs["LIC"]

plt.figure(figsize=(12,4),dpi=300,facecolor=(0.8,0.8,0.8))
plt.title("AUTOMOBILE ENGINE PISTON RINGS - CHART S",fontsize=12)
plt.xlabel("number of samples",fontsize =12)
plt.ylabel("diameter measures (mm)",fontsize=12)
plt.plot(x,y2,color=(0.137, 0.878, 0.369),lw=1.5,linestyle="--",label="UCL")
plt.plot(x,y3,color=(1, 0.643, 0.227),lw=1.5,linestyle="--",label="x̄")
plt.plot(x,y4,color=(0.071, 0.529, 0.961),lw=1.5,linestyle="--",label="LCL")
plt.plot(x,y1,color=(0.322, 0.333, 0.345),lw=1.5,linestyle="-",label="S",marker="o")
plt.legend(('UCL','CL','CLC','S'),prop={'size':10},loc=(1.02,0.4))
plt.tight_layout()
plt.grid(color=(0.322, 0.333, 0.345),linestyle='dashed',linewidth = 0.5)
plt.xticks(range(min(x), max(x)+1));
plt.figtext(0.780, 0.790, 'UCL=0.01964',fontsize=10)
plt.figtext(0.780, 0.23, 'LCL=0',fontsize=10)
plt.savefig('chart-s.png', format='png', dpi=1200)
plt.show()
```
