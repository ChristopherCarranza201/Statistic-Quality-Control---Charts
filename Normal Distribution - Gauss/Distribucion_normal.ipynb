```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
# constantes
mu=74
sigma=0.0084
x1=73.746
x2=74.254
```

```python
#Calculo z
z1=(x1-mu)/sigma
z2=(x2-mu)/sigma
print(z1,z2)
```

```python
#Graficar
x=np.arange(z1,z2,0.0001)
x2=np.arange(-10,10,0.0001)
y=norm.pdf(x,0,1)
y2=norm.pdf(x2,0,1)
fig,ax=plt.subplots(figsize=(10,5))
plt.style.use("fivethirtyeight")
ax.plot(x2,y2)
ax.fill_between(x,y,0,color="g")
ax.fill_between(x2,y2,0,alpha=0.1)
ax.set_xlim([-3,3])
ax.set_label("variaci√≥n")
ax.set_title("Campana de Gauss")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 74
sigma = 0.0084
n = 125

variaciones = np.random.normal(mu, sigma, n)
plt.figure(figsize=(9, 3.5))
conteo, x, barras = plt.hist(variaciones, 100, density=True, color="pink", alpha=0.7)

x_values = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
y_values = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x_values - mu) / sigma) ** 2)
plt.plot(x_values, y_values, linewidth=1, color="purple")

plt.axvline(mu - 3 * sigma, color="teal", linewidth=1.5, linestyle="dashed")
plt.axvline(mu - 2 * sigma, color="teal", linewidth=1.5, linestyle="dashed")
plt.axvline(mu - sigma, color="teal", linewidth=1.5, linestyle="dashed")
plt.axvline(mu, color="teal", linewidth=1.5, linestyle="dashed")
plt.axvline(mu + sigma, color="teal", linewidth=1.5, linestyle="dashed")
plt.axvline(mu + 2 * sigma, color="teal", linewidth=1.5, linestyle="dashed")
plt.axvline(mu + 3 * sigma, color="teal", linewidth=1.5, linestyle="dashed")

plt.yticks([])
plt.xticks([])
plt.box(False)
plt.show()
```

```python
print("\n*** Regla Emp√≠rica 68-95-99.7 de la Distribuci√≥n Normal ***\n")
print(mu-sigma,"-",mu+sigma,"cm\t",
      mu-2*sigma,"-",mu+2*sigma,"cm\t",
      mu-3*sigma,"-",mu+3*sigma,"cm")
print((np.where(((mu-sigma)<variaciones)&
                (variaciones<(mu+sigma)),1,0).sum()*100)/n,"%\t",(np.where(((mu-2*sigma)<variaciones)&
                (variaciones<(mu+2*sigma)),1,0).sum()*100)/n,"%\t",(np.where(((mu-3*sigma)<variaciones)&
                (variaciones<(mu+3*sigma)),1,0).sum()*100)/n,"%\t")
```
